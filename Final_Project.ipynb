<b> Next, we chose to look at poverty rates in all the neighborhoods. 

```python
data3 = pd.read_csv("data - Sheet1.tsv", usecols = ["Neighborhood","Est. Pop. Under Poverty (2010)",
                                                   "Edu. Attainment: Postgraduate Degree (2010)"], sep = '\\t' ,engine='python')
```

<b>We got the top ten highest poverty inflicted neighborhoods in Pittsburgh. The top 3 were Central Oakland, Shadyside. and East Liberty. The information shows that since these neighborhoods have such a high poverty rate, they cannot be in the best neighborhood category.


```python
data3["Est. Pop. Under Poverty (2010)"] = data3["Est. Pop. Under Poverty (2010)"].str.replace(",","").astype(float)
data5 = data3.sort_values("Est. Pop. Under Poverty (2010)", axis=0, 
                 ascending=False, inplace=False, kind='mergesort', na_position='last').head(10)
data5.plot.bar(x = 'Neighborhood', y = 'Est. Pop. Under Poverty (2010)')
# x-axix neighborhood y-aixs population
```

<b>as you can see from the graph, Central Oakland, Shadyside, and East liberty have a really high population rate and because of this, they cannot be chosen to be one of the best neighborhoods


<b>after the poverty rates, we looked at the percentage of people in the neighborhood that have received a postgraduate degree.


```python
data3["Edu. Attainment: Postgraduate Degree (2010)"] = data3["Edu. Attainment: Postgraduate Degree (2010)"].replace("%","").astype(float)
data4 = data3.sort_values("Edu. Attainment: Postgraduate Degree (2010)", axis=0, ascending=False, inplace=False, 
                         kind='mergesort', na_position='last').head(10)
data4.plot.bar(x = 'Neighborhood', y = 'Edu. Attainment: Postgraduate Degree (2010)')
# x-axis neighborhood y-axis postgraduate degree percent
```

<b>as show from the bar graph, South Shore, and regent square dominate in that aspect. the graph also correlates with the poverty bar graph because none of the neighborhoods that have a high poverty rate have a high percentage of people with a post secondary degree
